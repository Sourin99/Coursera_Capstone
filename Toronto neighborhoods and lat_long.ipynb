Reading the data

```python
df_n = pd.read_csv('/kaggle/input/Toronto_Neighborhood_data.csv')
df_c = pd.read_csv('/kaggle/input/Geospatial_Coordinates.csv')
```

Check if both the data sets match

```python
df_n.size == df_c.size
```

Sorting both the datasets to make copying lat,long easier

```python
df_n.sort_values(by='Postal Code',axis=0,inplace=True)
df_c.sort_values(by='Postal Code',axis=0,inplace=True)

df_n.reset_index(inplace=True)
df_c.reset_index(inplace=True)
```

Append Lat,Long directly into the other table

```python
df_n['Latitude'] = df_c['Latitude']
df_n['Longitude'] = df_c['Longitude']
```

```python
df_n.drop(['index'],axis=1,inplace=True)
df_n
```

Exporting the dataframe to a csv and storing it

```python
df_n.to_csv('Toronto_Postal_data.csv',index=False)
```
